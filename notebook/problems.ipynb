## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Write the above functions but for both normal and uniform distributions.

```python
from scipy.stats import norm

# Define todas las funciones de probabilidad relacionadas con esta distribución

def dnorm(x, mean=0, std=1):
    """
    Calcula la densidad de la distribución normal.
    """
    result = norm.pdf(x, loc=mean, scale=std)
    return result

def pnorm(x, mean=0, std=1):
    """
    Calcula la acumulativa de la distribución normal.
    """
    result = norm.cdf(x=x, loc=mean, scale=std)
    return result

def qnorm(p, mean=0, std=1):
    """
    Calcula la función cuantil de la distribución normal.
    """
    result = norm.ppf(q=p, loc=mean, scale=std)
    return result

def rnorm(n, mean=0, std=1):
    """
    Genera variables aleatorias de la distribución normal.
    """
    result = norm.rvs(size=n, loc=mean, scale=std)
    return result

# Ejemplos de uso de las funciones
print("Densidad de la distribución normal para x=0:", dnorm(0))
print("Cumulativa de la distribución normal para x=0:", pnorm(0))
print("Quantil de la distribución normal para p=0.5:", qnorm(0.5))
print("Generar 5 variables aleatorias de una distribución normal:", rnorm(5))
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

mean = 0
std = 1
data = np.arange(-5, 5, 0.01)

pmf = dnorm(data, mean = mean, std = std)
cdf = pnorm(data, mean = mean, std = std)
ppf = qnorm(data, mean = mean, std = std)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(data, pmf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(data, ppf, "g-")

# Ajuste del diseño 
plt.tight_layout()

plt.show()
```

```python
from scipy.stats import uniform
import numpy as np

def dunif(x, start=0, width=1):
    """
    Calculates the density of the uniform distribution at a specific point.
    Uniform density is constant across [start, start + width].
    """
    return uniform.pdf(x, loc=start, scale=width)

def punif(q, start=0, width=1):
    """
    Calculates the cumulative probability of the uniform distribution up to q.
    """
    return uniform.cdf(q, loc=start, scale=width)

def qunif(p, start=0, width=1):
    """
    Calculates the quantile function from the uniform distribution for a given cumulative probability p.
    """
    return uniform.ppf(p, loc=start, scale=width)

def runif(n, start=0, width=1):
    """
    Generates random variables from the uniform distribution.
    """
    return uniform.rvs(loc=start, scale=width, size=n)

# Ejemplo de uso
np.random.seed(42)

print(f"Density of the Uniform at point 0.5 (should be constant for [0,1]): {dunif(0.5, start=0, width=1)}")
print(f"Cumulative probability up to 0.5 for Uniform(0,1): {punif(0.5, start=0, width=1)}")
print(f"Which value has a cumulative probability of 0.9 in Uniform(0,1)?: {qunif(0.9, start=0, width=1)}")
print(f"Generate 10 random values from a Uniform(0,1) distribution: {runif(10, start=0, width=1)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

def d_uniforme(x, bajo=0, alto=1):
    """
    Calcula la densidad de la distribución uniforme en un punto específico.
    """
    return uniform.pdf(x, loc=bajo, scale=(alto - bajo))

def p_uniforme(q, bajo=0, alto=1):
    """
    Calcula la probabilidad acumulativa de la distribución uniforme hasta un punto q.
    """
    return uniform.cdf(q, loc=bajo, scale=(alto - bajo))

def q_uniforme(p, bajo=0, alto=1):
    """
    Calcula la función cuantil de la distribución uniforme para una probabilidad acumulativa dada p.
    """
    return uniform.ppf(p, loc=bajo, scale=(alto - bajo))

def r_uniforme(n, bajo=0, alto=1):
    """
    Genera valores aleatorios de la distribución uniforme.
    """
    return uniform.rvs(loc=bajo, scale=(alto - bajo), size=n)

# Configurar la semilla aleatoria para reproducibilidad
np.random.seed(42)

# Parámetros de la distribución
bajo = 0
alto = 1
datos = np.linspace(bajo - 1, alto + 1, 400)  # Rango extendido para visualización

# Calcular métricas
pmf = d_uniforme(datos, bajo=bajo, alto=alto)
cdf = p_uniforme(datos, bajo=bajo, alto=alto)
ppf = q_uniforme(np.linspace(0, 1, 400), bajo=bajo, alto=alto)  # Valores de p de 0 a 1

# Visualización
fig, axis = plt.subplots(3, 1, figsize=(5, 7))
axis[0].plot(datos, pmf, "r-", label='Densidad')
axis[1].plot(datos, cdf, "b-", label='Acumulativa')
axis[2].plot(np.linspace(0, 1, 400), ppf, "g-", label='Cuantil')

# Ajustar la disposición de los gráficos
for ax in axis:
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```
